```python
!pip install cupy-cuda12x -q
import numpy as np
import cupy as cp
import matplotlib.pyplot as plt
import time
import random
from concurrent.futures import ProcessPoolExecutor
import multiprocessing as mp

print(f"GPU: {cp.cuda.is_available()}, CPU cores: {mp.cpu_count()}")
```

```python
def generate_cities(seed, n, width=100, height=100):
    random.seed(seed)
    return np.array([(random.randint(0, width), random.randint(0, height)) for _ in range(n)])

def distance_matrix_numpy(cities):
    n = len(cities)
    cities = cities.astype(np.float32)
    diff = cities[:, np.newaxis, :] - cities[np.newaxis, :, :]
    return np.sqrt(np.sum(diff**2, axis=2))

# def path_length(path, dist):
#     path_t = 0
#     for i in range(len(path) - 1):
#         if path_t:
#           path_t += dist[path[i], path[i+1]]
#     path_h += dist[path[-1], path[0]]
#     return path_t

def path_length_numpy(path, dist_matrix):
    total = 0
    for i in range(len(path) - 1):
        total += dist_matrix[path[i], path[i+1]]
    total += dist_matrix[path[-1], path[0]]
    return total

def mutate(path):
    path = path.copy()
    a, b = random.sample(range(len(path)), 2)
    path[a], path[b] = path[b], path[a]
    return path

def crossover(p1, p2):
    cut = random.randint(1, len(p1)-2)
    return np.concatenate([p1[:cut], [c for c in p2 if c not in p1[:cut]]])
```

```python
def ga_sequential(cities, pop_size, generations, seed):
    print("seq")
    random.seed(seed)
    np.random.seed(seed)
    n = len(cities)
    dist_matrix = distance_matrix_numpy(cities)

    population = [np.random.permutation(n) for _ in range(pop_size)]
    history = []
    start_time = time.time()

    for gen in range(generations):
        fitnesses = []
        for individual in population:
            fitness = path_length_numpy(individual, dist_matrix)
            fitnesses.append(fitness)
        fitnesses = np.array(fitnesses)

        best_fitness = np.min(fitnesses)
        history.append(best_fitness)

        sorted_indices = np.argsort(fitnesses)
        elite_size = max(2, pop_size // 10)
        elite_indices = sorted_indices[:pop_size // 5]

        new_population = [population[sorted_indices[i]].copy() for i in range(elite_size)]

        while len(new_population) < pop_size:
            p1_idx, p2_idx = np.random.choice(elite_indices, 2, replace=False)
            child = mutate(crossover(population[p1_idx], population[p2_idx]))
            new_population.append(child)

        population = new_population

        if (gen+1) % 100 == 0:
            print(f"Gen {gen+1} | Best: {best_fitness:.2f} | Time: {time.time()-start_time:.2f}s")

    elapsed = time.time() - start_time
    best_idx = np.argmin(fitnesses)
    best = population[best_idx]
    best_dist = fitnesses[best_idx]

    print(f"FINAL: {best_dist:.2f} in {elapsed:.2f}s\n")
    return best, best_dist, elapsed, history
```

```python
def evaluate_batch(args):
    individuals, dist_matrix, start_idx = args
    results = []
    for i, individual in enumerate(individuals):
        fitness = path_length_numpy(individual, dist_matrix)
        results.append((start_idx + i, fitness))
    return results

def create_offspring_batch_parallel(args):
    elite_pool, batch_size, seed_offset = args
    random.seed(seed_offset)
    np.random.seed(seed_offset)
    offspring = []
    for _ in range(batch_size):
        p1_idx, p2_idx = np.random.choice(len(elite_pool), 2, replace=False)
        child = mutate(crossover(elite_pool[p1_idx], elite_pool[p2_idx]))
        offspring.append(child)
    return offspring

def ga_parallel(cities, pop_size, generations, n_workers, seed):
    random.seed(seed)
    np.random.seed(seed)
    n = len(cities)
    dist_matrix = distance_matrix_numpy(cities)

    population = [np.random.permutation(n) for _ in range(pop_size)]
    history = []
    start_time = time.time()

    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        for gen in range(generations):
            batch_size = pop_size // n_workers
            tasks = []

            for i in range(n_workers):
                start = i * batch_size
                end = start + batch_size if i < n_workers - 1 else pop_size
                batch = population[start:end]
                tasks.append((batch, dist_matrix, start))

            results = executor.map(evaluate_batch, tasks)

            fitnesses = np.zeros(pop_size)
            for batch_results in results:
                for idx, fitness in batch_results:
                    fitnesses[idx] = fitness

            best_fitness = np.min(fitnesses)
            history.append(best_fitness)

            sorted_indices = np.argsort(fitnesses)
            elite_size = max(2, pop_size // 10)
            elite_pool = [population[i] for i in sorted_indices[:pop_size // 5]]

            new_population = [population[sorted_indices[i]].copy() for i in range(elite_size)]

            num_offspring = pop_size - elite_size
            batch_size_genetic = num_offspring // n_workers
            tasks_genetic = []

            for i in range(n_workers):
                size = batch_size_genetic + (1 if i < num_offspring % n_workers else 0)
                tasks_genetic.append((elite_pool, size, seed + gen * 1000 + i))

            offspring_batches = executor.map(create_offspring_batch_parallel, tasks_genetic)
            for batch in offspring_batches:
                new_population.extend(batch)

            population = new_population

            if (gen+1) % 100 == 0:
                print(f"[PAR] Gen {gen+1} | Best: {best_fitness:.2f} | Time: {time.time()-start_time:.2f}s")

    elapsed = time.time() - start_time
    best_idx = np.argmin(fitnesses)
    best = population[best_idx]
    best_dist = fitnesses[best_idx]

    print(f"[PAR] FINAL: {best_dist:.2f} in {elapsed:.2f}s\n")
    return best, best_dist, elapsed, history
```

```python
evaluate_paths_kernel = cp.RawKernel(r'''
extern "C" __global__
void evaluate_paths(const int* population, const float* dist_matrix,
                   float* fitness, int pop_size, int n_cities) {
    int idx = blockDim.x * blockIdx.x + threadIdx.x;

    if (idx < pop_size) {
        float total = 0.0f;

        for (int i = 0; i < n_cities - 1; i++) {
            int from = population[idx * n_cities + i];
            int to = population[idx * n_cities + i + 1];
            total += dist_matrix[from * n_cities + to];
        }

        int last = population[idx * n_cities + (n_cities - 1)];
        int first = population[idx * n_cities];
        total += dist_matrix[last * n_cities + first];

        fitness[idx] = total;
    }
}
''', 'evaluate_paths')
```

```python
def ga_gpu(cities, pop_size, generations, seed):
    print("gpu")
    random.seed(seed)
    np.random.seed(seed)
    n = len(cities)

    dist_matrix_cpu = distance_matrix_numpy(cities).astype(np.float32)
    dist_matrix_gpu = cp.asarray(dist_matrix_cpu)

    population_cpu = [np.random.permutation(n) for _ in range(pop_size)]
    history = []
    start_time = time.time()

    threads_per_block = 256
    blocks = (pop_size + threads_per_block - 1) // threads_per_block

    for gen in range(generations):
        pop_array = np.array(population_cpu, dtype=np.int32)
        population_gpu = cp.asarray(pop_array)
        fitness_gpu = cp.zeros(pop_size, dtype=cp.float32)

        evaluate_paths_kernel(
            (blocks,), (threads_per_block,),
            (population_gpu, dist_matrix_gpu, fitness_gpu, pop_size, n)
        )

        fitnesses = cp.asnumpy(fitness_gpu)
        # +_

        best_fitness = np.min(fitnesses)
        history.append(best_fitness)
        sorted_indices = np.argsort(fitnesses)
        elite_size = max(2, pop_size // 10)
        elite_indices = sorted_indices[:pop_size // 5]

        new_population = [population_cpu[sorted_indices[i]].copy() for i in range(elite_size)]

        while len(new_population) < pop_size:
            p1_idx, p2_idx = np.random.choice(elite_indices, 2, replace=False)
            child = mutate(crossover(population_cpu[p1_idx], population_cpu[p2_idx]))
            new_population.append(child)

        population_cpu = new_population

        if (gen+1) % 100 == 0:
            print(f"Gen {gen+1} | Best: {best_fitness:.2f} | Time: {time.time()-start_time:.2f}s")

    elapsed = time.time() - start_time
    best_idx = np.argmin(fitnesses)
    best = population_cpu[best_idx]
    best_dist = fitnesses[best_idx]

    print(f"FINAL: {best_dist:.2f} in {elapsed:.2f}s\n")
    return best, best_dist, elapsed, history
```

```python
def plot_comp(cities, results_seq, results_par, results_gpu):
    best_seq, dist_seq, time_seq, history_seq = results_seq
    best_par, dist_par, time_par, history_par = results_par
    best_gpu, dist_gpu, time_gpu, history_gpu = results_gpu

    fig1, ax = plt.subplots(figsize=(10, 8))
    path = np.concatenate([best_seq, [best_seq[0]]])
    ax.plot(cities[path, 0], cities[path, 1], '-o', color='blue', linewidth=2, markersize=6)
    ax.scatter(cities[:, 0], cities[:, 1], c='blue', s=100, zorder=5, alpha=0.7)
    ax.set_title(f'TSP \dist: {dist_seq:.2f}', fontsize=14, fontweight='bold')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    fig2, ax = plt.subplots(figsize=(10, 6))
    methods = ['seq', 'Parallel\n(CPU)', 'GPU']
    times = [time_seq, time_par, time_gpu]
    colors = ['blue', 'orange', 'green']
    bars = ax.bar(methods, times, color=colors, alpha=0.7, width=0.6)
    for bar, t in zip(bars, times):
        ax.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                f'{t:.2f}s', ha='center', va='bottom', fontweight='bold', fontsize=12)
    ax.set_ylabel('t')
    ax.set_title('czas', fontweight='bold')
    ax.grid(True, axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

    speedup_par = time_seq / time_par
    speedup_gpu = time_seq / time_gpu

    print(f"seq:{time_seq:.2f}s | dist: {dist_seq:.2f}")
    print(f"par:{time_par:.2f}s | dist: {dist_par:.2f} | Speedup: {speedup_par:.2f}x")
    print(f"GPU:{time_gpu:.2f}s | dist: {dist_gpu:.2f} | Speedup: {speedup_gpu:.2f}x")
```

```python
SEED = 42
NUM_CITIES = 100
POP_SIZE = 2500
GENERATIONS = 500
N_WORKERS = mp.cpu_count()

print(f"cities: {NUM_CITIES}, pop: {POP_SIZE}, gens: {GENERATIONS}\n")

cities = generate_cities(SEED, NUM_CITIES)

results_seq = ga_sequential(cities, POP_SIZE, GENERATIONS, SEED)
results_par = ga_parallel(cities, POP_SIZE, GENERATIONS, N_WORKERS, SEED)
results_gpu = ga_gpu(cities, POP_SIZE, GENERATIONS, SEED)

plot_comp(cities, results_seq, results_par, results_gpu)
```

```python
def create_offspring_batch(args):
    elite_pool, batch_size, seed_offset = args
    random.seed(seed_offset)
    np.random.seed(seed_offset)
    offspring = []
    for _ in range(batch_size):
        p1_idx, p2_idx = np.random.choice(len(elite_pool), 2, replace=False)
        child = mutate(crossover(elite_pool[p1_idx], elite_pool[p2_idx]))
        offspring.append(child)
    return offspring

def ga_parallel_gpu(cities, pop_size, generations, n_workers, seed):
    random.seed(seed)
    np.random.seed(seed)
    n = len(cities)

    dist_matrix_cpu = distance_matrix_numpy(cities).astype(np.float32)
    dist_matrix_gpu = cp.asarray(dist_matrix_cpu)

    population_cpu = [np.random.permutation(n) for _ in range(pop_size)]
    history = []
    start_time = time.time()

    threads_per_block = 256
    blocks = (pop_size + threads_per_block - 1) // threads_per_block

    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        for gen in range(generations):
            # ========== GPU FITNESS EVALUATION ==========
            pop_array = np.array(population_cpu, dtype=np.int32)
            population_gpu = cp.asarray(pop_array)
            fitness_gpu = cp.zeros(pop_size, dtype=cp.float32)

            evaluate_paths_kernel(
                (blocks,), (threads_per_block,),
                (population_gpu, dist_matrix_gpu, fitness_gpu, pop_size, n)
            )

            fitnesses = cp.asnumpy(fitness_gpu)
            # ============================================

            best_fitness = np.min(fitnesses)
            history.append(best_fitness)

            # ========== PARALLEL GENETIC OPERATIONS ==========
            sorted_indices = np.argsort(fitnesses)
            elite_size = max(2, pop_size // 10)
            elite_pool = [population_cpu[i] for i in sorted_indices[:pop_size // 5]]

            new_population = [population_cpu[sorted_indices[i]].copy() for i in range(elite_size)]

            num_offspring = pop_size - elite_size
            batch_size = num_offspring // n_workers
            tasks = []
            for i in range(n_workers):
                size = batch_size + (1 if i < num_offspring % n_workers else 0)
                tasks.append((elite_pool, size, seed + gen * 1000 + i))

            offspring_batches = executor.map(create_offspring_batch, tasks)
            for batch in offspring_batches:
                new_population.extend(batch)

            population_cpu = new_population
            # =================================================

            if (gen+1) % 100 == 0:
                print(f"{gen+1} | f: {best_fitness:.2f} | t: {time.time()-start_time:.2f}s")

    elapsed = time.time() - start_time
    best_idx = np.argmin(fitnesses)
    best = population_cpu[best_idx]
    best_dist = fitnesses[best_idx]

    print(f"[PAR+GPU] FINAL: {best_dist:.2f} in {elapsed:.2f}s\n")
    return best, best_dist, elapsed, history

print("PARALLEL + GPU\n")
result_combined = ga_parallel_gpu(cities, POP_SIZE, GENERATIONS, N_WORKERS, SEED)

print(f"{result_combined[2]:.2f}s, distance {result_combined[1]:.2f}")
```
